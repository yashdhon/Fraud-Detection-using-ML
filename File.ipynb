```python
import numpy as np
import pandas as pd
```

```python
fraud=pd.read_csv("/content/Fraud.csv", engine='python', on_bad_lines='skip')
fraud.head()
```

```python
fraud['oldbalanceOrg'] = pd.to_numeric(fraud['oldbalanceOrg'], errors='coerce')
fraud['newbalanceOrig'] = pd.to_numeric(fraud['newbalanceOrig'], errors='coerce')

```

```python

fraud['amount'] = pd.to_numeric(fraud['amount'], errors='coerce')


fraud['delta_orig'] = fraud['oldbalanceOrg'] - fraud['newbalanceOrig']
fraud['delta_dest'] = fraud['newbalanceDest'] - fraud['oldbalanceDest']

fraud['rel_amount'] = fraud['amount'] / (fraud['oldbalanceOrg'] + 1)  # avoid division by zero
```

```python
fraud = pd.get_dummies(fraud, columns=['type'])
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()



fraud['type_CASH_IN']=le.fit_transform(fraud['type_CASH_IN'])
fraud['type_CASH_OUT']=le.fit_transform(fraud['type_CASH_OUT'])
fraud['type_DEBIT']=le.fit_transform(fraud['type_DEBIT'])

```

```python
fraud['type_PAYMENT']=le.fit_transform(fraud['type_PAYMENT'])
fraud['type_TRANSFER']=le.fit_transform(fraud['type_TRANSFER'])
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
```

```python
clf = Pipeline(steps=[

    ('classifier', RandomForestClassifier(
        n_estimators=200, random_state=42, class_weight='balanced'))
])
```

```python
x=fraud.iloc[:,[0,1,3,4,6,7,9,10,11,12,13,14,15,16,17]]
```

```python
y=fraud.iloc[:,8]
```

```python
x
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42, stratify=y)

clf.fit(x_train, y_train)
```

```python
y_pred=clf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print("test accu",accuracy_score(y_test,y_pred))
print("train accu",accuracy_score(y_train,clf.predict(x_train)))
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```
